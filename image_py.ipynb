```python
# from google.colab import files
# uploaded = files.upload()

# Image="/content/20240330_082324-COLLAGE.jpg"

from PIL import Image # Changed 'image' to 'Image'olt
import matplotlib.pyplot as plt

image = Image.open("/content/WhatsApp_Image_2024-06-29_at_9.55.21_AM-removebg-preview (1).jpg")
# image = files.upload()
image

# import cv2

import cv2
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

image = Image.open("/content/WhatsApp_Image_2024-06-29_at_9.55.21_AM-removebg-preview (1).jpg")
# image = files.upload()
# Convert the PIL Image object to a NumPy array
image_np = np.array(image)
# Convert the NumPy array from RGB to BGR
# (OpenCV uses BGR color format by default)
image_bgr = cv2.cvtColor(image_np, cv2.COLOR_RGB2BGR)
# Now you can convert to grayscale
grayed = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

plt.imshow(grayed, cmap='gray')
plt.show()

inverted = cv2.bitwise_not(grayed)
plt.imshow(inverted,cmap='gray')

blurred = cv2.GaussianBlur(inverted, (21, 21), 0)
plt.imshow(blurred,cmap='gray')

def blend(x, y):
    return cv2.divide(x, 255 - y,scale=256)

final_result = blend(grayed, blurred)

cv2.imwrite("my_artwork.jpg", final_result)

plt.imshow(final_result,cmap='gray')


```
